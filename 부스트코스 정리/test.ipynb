```python
import pymysql
import pandas as pd
import matplotlib.pyplot as plt
conn,cur = None,None
data1,data2,data3,data4 = "","","",""
sql = ""
conn = pymysql.connect(host='127.0.0.1',user='root',password='0000',db='stardb',charset='utf8')
cur = conn.cursor()
cur.execute("use practice;")
sql = "select 연령대 ,count(연령대) as 연령대별 from customer_profile group by 연령대 order by 연령대 asc;"
cur.execute(sql)
result = cur.fetchall()
result = pd.DataFrame(result)


df = pd.read_sql(sql, conn)
print(df)
plt.rc('font', family='Malgun Gothic')
c_prc = df['연령대별']
c_prc.index = df['연령대']
plt.figure(figsize=(11,9))
c_prc.plot(label='연령대별 수량', title= "선호도 조사")
plt.legend(loc='lower left')
plt.grid(True)
plt.show()

```
